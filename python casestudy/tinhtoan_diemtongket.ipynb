# <h1>Case study: Bài toán phân tích bảng điểm học sinh</h1>

<b>Tổng quan dự án</b>

Cho thông tin file bảng điểm chi tiết của các học sinh trong lớp các môn học liên quan đến việc thi đại học như sau: Toán, Lý, Hóa, Sinh, Văn, Anh, Sử, Địa. Mỗi một môn học tự nhiên có 4 đầu điểm: Điểm kiểm tra miệng, kiểm tra 15 phút, kiểm tra 1 tiết, thi cuối kỳ (5/10/15/70). Mỗi môn học xã hội có 5 đầu điểm: Điểm chuyên cần, điểm bài luận, kiểm tra 1 tiết, kiểm tra 2 tiết, thi cuối kỳ (5, 10, 10, 15, 60).

Hãy viết một chương trình thực hiện các nhiệm vụ sau:

- Load dữ liệu bảng điểm chi tiết vào.

- Tính điểm tổng kết trung bình cho từng môn.

- Xếp loại học lực của học sinh (trung bình, trung bình khá, khá, giỏi dựa vào điểm tổng kết trung bình của học sinh).

- Hàng đầu tiên gồm các đề mục: "Mã HS, Toán , Lý, Hóa, Sinh, Văn, Anh, Sử, Địa". Hàng thứ 2 trở đi là bảng điểm chi tiết cho từng sinh viên (tên sinh viên + điểm chi tiết). Mỗi môn học tự nhiên có 4 đầu điểm, mỗi môn học xã hội có 5 đầu điểm. Các điểm thành phần của 1 môn được phân cách bằng dấu phẩy, các môn được phân cách bằng dấu chấm phẩy.

- Tỉ lệ điểm thành phần cho các môn tự nhiên (Toán, Lý, Hóa, Sinh): Điểm kiểm tra miệng, kiểm tra 15 phút, kiểm tra 1 tiết, thi cuối kỳ (5%, 10%, 15%, 70%).

- Tỉ lệ điểm thành phần cho các môn xã hội (Anh, Văn, Sử, Địa): Điểm chuyên cần, điểm bài luận, kiểm tra 1 tiết, kiểm tra 2 tiết, thi cuối kỳ (5%, 10%, 10%, 15%, 60%).

- Điểm trung bình của từng môn sẽ được tính toán theo tỉ lệ điểm thành phần được quy định ở trên, kết quả làm tròn đến 2 chữ số. VD: Diem_TB_Toan_SVA = 5% * kiem_tra_mieng + 10% * kiem_tra_15 + 15% * kiem_tra_1tiet + 70% * thi_cuoi_ky.

```python
# giao diện cho case study

from tkinter import *

sc = Tk()

lb_tieude = Label(sc, text = 'PHÂN TÍCH ĐIỂM HỌC SINH', fg = 'red', font = ("Arial",22))
link_dct = StringVar()

# Khởi tạo Hàm tính điểm trung bình
def tinhdiem_trungbinh(link_dct):
    kq = ""
    # Đọc dữ liệu từ file điểm chi tiết
    hand = open(link_dct, "r", encoding = "utf8")

    # Đọc từng dòng dữ liệu trong file
    lines = hand.readlines()

    # Khai báo các list cho môn học, học sinh, bảng điểm
    monhoc = []
    hocsinh = []
    bangdiem = []

    # Thêm dữ liệu vào list monhoc
    a = lines[0].strip()
    word = a.split(',')
    for char in word:
        monhoc.append(char)

    monhoc = monhoc[1:]

    # Thêm dữ liệu vào list hocsinh, bangdiem
    for line in lines[1:]:
        sline = line.strip()
        sline = sline.split(';')
        hocsinh.append(sline[0])
        bangdiem.append(sline[1:])

    # khai báo biến global dùng cho hàm lưu điểm trung bình
    global hocsinh2        
    hocsinh2 = hocsinh

    # Khởi tạo hàm tính điểm trung bình môn tự nhiên 5%, 10%, 15%, 70%
    def diemTB_monhocTN(x):
        return 0.05*x[0] + 0.1*x[1] + 0.15*x[2] + 0.7*x[3]

    # khởi tạo hàm tính điểm trung bình môn xã hội    5%, 10%, 10%, 15%, 60%
    def diemTB_monhocXH(y):
        return 0.05*y[0] + 0.1*y[1] + 0.1*y[2] + 0.15*y[3] + 0.6*y[4]

    # Tạo list bangdiem_update để lưu điểm các môn học theo dạng số thực
    bangdiem_update = []
    for i in bangdiem:
        for j in i:
            j = j.replace(" ", "")
            k = j.split(',')
            h = list(map(float,k))
            bangdiem_update.append(h)

    # tạo list bangdiem_TB  để lưu dữ liệu điểm trung bình
    bangdiem_TB = []
    for i in bangdiem_update:
        if (len(i) == 4):
            a = round(diemTB_monhocTN(i), 2)        
            bangdiem_TB.append(a)
        if (len(i) == 5):
            b = round(diemTB_monhocXH(i), 2)
            bangdiem_TB.append(b)  

    # Chia list bangdiem_TB thành 6 list con bên trong nó
    bangdiem_TB = [bangdiem_TB[i:i + 8] for i in range(0, len(bangdiem_TB), 8)]

    # Khai báo biến global diemTB dùng hoc hàm lưu điểm trung bình
    global diemTB
    diemTB = bangdiem_TB    

    # Tạo dict diemMH để gộp dữ liệu của 2 list monhoc và bangdiem_TB
    diemMH = []
    for i in bangdiem_TB:
        x = zip(monhoc, i)
        x = dict(x)
        diemMH.append(x)

    # Tạo dict tongket_MH để gộp dữ liệu của 2 list hocsinh và diemMH
    tongket_MH = zip(hocsinh, diemMH)
    tongket_MH = dict(tongket_MH)
    
    # In dict tongket_MH ra màn hình
    for key in tongket_MH:
        kq += (str(key) +  ":" + str(tongket_MH[key])+ "\n")
    # Đóng file diem_chitiet.txt
    hand.close()
    return kq

# Khởi tạo Hàm lưu điểm trung bình
def luudiem_trungbinh(hocsinh2, diemTB, link_dtb):

    # Gộp danh sách hocsinh va diemTB vào danh sách x
    x = zip(hocsinh2, diemTB)
    x = list(x)

    # Đọc file diem_trungbih.txt để ghi dữ liệu
    f = open(link_dtb, "w", encoding = "utf8")

    f.write('Mã HS, Toán, Lý, Hóa, Sinh, Văn, Anh, Sử, Địa' + '\n')

    # Lấy dữ liệu của list x để ghi ra file
    for i in x:
        i = str(i)
        i = i.replace("(","")
        i = i.replace(")","")
        i = i.replace("[","")
        i = i.replace("]","")
        i = i.replace(",",";")
        i = i.replace("'","")
        f.write(i)
        f.write("\n")
    f.close()

# khởi tạo hàm xeploai_hocsinnh
def xeploai_hocsinh(link_dtb): 

    # Mở file diem_trungbinh.txt để đọc dữ liệu
    data = open(link_dtb, "r", encoding = "utf8")

    # Đọc từng dòng dữ liệu trong file
    lines = data.readlines()

    # Khai báo các list cho học sinh, bảng điểm
    hocsinh = []
    bangdiem = []

    # Thêm dữ liệu vào list hocsinh, bangdiem
    for line in lines[1:]:
        sline = line.strip()
        sline = sline.split(';')
        hocsinh.append(sline[0])
        bangdiem.append(sline[1:])

    # chuyển list bangdiem sang list bangdiem_update với định dạng số thực
    bangdiem_update = []
    for i in bangdiem:
        for j in i:
            j = j.replace(" ", "")
            k = j.split(',')
            h = list(map(float,k))
            for v in h:
                bangdiem_update.append(v)

    # Chia list bangdiem_update thành 6 list con
    bangdiem_update = [bangdiem_update[i:i+8] for i in range(0, len(bangdiem_update), 8)]

    #  Hàm Tính điểm trung bình chuẩn
    def dtb_chuan(x):
        return ((x[0] + x[4] + x[5])*2.0 + (x[1] + x[2] + x[3] + x[6] + x[7])*1.0)/11.0

    # tạo danh sách indiemTB để lưu dữ liệu sau khi tính trung bình chuẩn
    indiemTB = []
    for y in bangdiem_update:
        a = round(dtb_chuan(y), 2)
        indiemTB.append(a)

    # tạo list tonghopDiem để gộp dữ liệu từ 2 list indiemTB và bangdiem_update
    tonghopDiem = zip(indiemTB, bangdiem_update)
    tonghopDiem = list(tonghopDiem)

    # Tạo list xeploai để đánh giá điểm của học sinh
    xeploai = []
    for i in tonghopDiem:
        if (i[0] >= 9.0 and min(i[1]) >=8):
            xl = "xuat sac"
        elif (i[0] >= 8.0 and min(i[1]) >= 6.5):
            xl = "gioi"
        elif (i[0] >= 6.5 and min(i[1]) >= 5.0):
            xl = "kha"
        elif (i[0] >= 6.0 and min(i[1]) >= 4.5):
            xl = "TB kha"
        else:
            xl = "TB"
        xeploai.append(xl)

    # Tổng hợp xếp loại học sinh và in ra màn hình
    b = zip(hocsinh, xeploai)
    b = dict(b)
    
    #đóng file diem_trungbinh.txt
    data.close()
    return b
    
def thuchien():
    chuoi_kq = ''
    try:
        link_dct.get()
        chuoi_kq += str(tinhdiem_trungbinh(link_dct.get()))    
        luudiem_trungbinh(hocsinh2, diemTB, "diem_trungbinh.txt")
        chuoi_kq += "\n" + str(xeploai_hocsinh("diem_trungbinh.txt"))
    except FileNotFoundError:
        chuoi_kq += 'File không tồn tại!'
    
    hienthi(chuoi_kq)
    
def hienthi(chuoi_kq):    
    lb_ketqua.config(text = chuoi_kq)

def thoat():
    sc.destroy()
    
lb_duongdan = Label(sc, text = "Nhập đường dẫn file", font = ("Arial",12))
en_diemct = Entry(sc,font = ("Arial",12), width = 55, textvariable = link_dct)

bt_thuchien = Button(sc,text = "Thực hiện",fg = 'blue',font = ("Arial",22), width = 10, command = thuchien)
bt_thoat = Button(sc,text = "Thoát",fg = 'red',font = ("Arial",22), width = 10, command = thoat)

lb_ketqua = Label(sc,text = "",justify = "left",bg = 'white',relief = 'sunken', width = 120, height = 20,anchor = "nw",font = ("Courier New",12))

## bố trí layout
lb_tieude.grid(column = 1, row = 1, columnspan = 2, pady = (10,40), padx = (40,40))
lb_duongdan.grid(column = 1, row = 2, columnspan = 2)
en_diemct.grid(column = 1, row = 3,columnspan = 2, padx = (40, 40), pady = (20, 20))

bt_thuchien.grid(column = 1, row = 4)
bt_thoat.grid(column = 2, row = 4)

lb_ketqua.grid(column = 1, row = 5, columnspan = 2, pady = (50,50), padx = (40, 40))

sc.mainloop()
```
