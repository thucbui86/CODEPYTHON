```python
# khởi tạo hàm xeploai_hocsinnh
def xeploai_hocsinh(link_dtb): 
    
    # Mở file diem_trungbinh.txt để đọc dữ liệu
    data = open(link_dtb, "r")

    # Đọc từng dòng dữ liệu trong file
    lines = data.readlines()
    
    # Khai báo các list cho học sinh, bảng điểm
    hocsinh = []
    bangdiem = []

    # Thêm dữ liệu vào list hocsinh, bangdiem
    for line in lines[1:]:
        sline = line.strip()
        sline = sline.split(';')
        hocsinh.append(sline[0])
        bangdiem.append(sline[1:])
    print(hocsinh)
    print(bangdiem)
    # khởi tạo biến global hocsinh2 để dùng cho hàm xếp loại học sinh thi đại học
    global hocsinh2
    hocsinh2 = hocsinh
    
    # chuyển list bangdiem sang list bangdiem_update với định dạng số thực
    bangdiem_update = []
    for i in bangdiem:
        for j in i:
            j = j.replace(" ", "")
            k = j.split(',')
            h = list(map(float,k))
            for v in h:
                bangdiem_update.append(v)
                
    # Chia list bangdiem_update thành 6 list con
    bangdiem_update = [bangdiem_update[i:i+8] for i in range(0, len(bangdiem_update), 8)]
    
    # Khởi tạo biến global bangdiem_danhgia dùng cho hàm xếp loại học sinh thi đại học
    global bangdiem_danhgia
    bangdiem_danhgia = bangdiem_update
    
    #  Hàm Tính điểm trung bình chuẩn
    def dtb_chuan(x):
        return ((x[0] + x[4] + x[5])*2.0 + (x[1] + x[2] + x[3] + x[6] + x[7])*1.0)/11.0
        
    # tạo danh sách indiemTB để lưu dữ liệu sau khi tính trung bình chuẩn
    indiemTB = []
    for y in bangdiem_update:
        a = round(dtb_chuan(y), 2)
        indiemTB.append(a)
        
    # tạo list tonghopDiem để gộp dữ liệu từ 2 list indiemTB và bangdiem_update
    tonghopDiem = zip(indiemTB, bangdiem_update)
    tonghopDiem = list(tonghopDiem)
    
    # Tạo list xeploai để đánh giá điểm của học sinh
    xeploai = []
    for i in tonghopDiem:
        if (i[0] >= 9.0 and min(i[1]) >=8):
            xl = "xuat sac"
        elif (i[0] >= 8.0 and min(i[1]) >= 6.5):
            xl = "gioi"
        elif (i[0] >= 6.5 and min(i[1]) >= 5.0):
            xl = "kha"
        elif (i[0] >= 6.0 and min(i[1]) >= 4.5):
            xl = "TB kha"
        else:
            xl = "TB"
        xeploai.append(xl)
        
    # Khởi tạo list phanloai dạng global để dùng cho hàm xeploai_hocsinh_thidaihoc()
    global phanloai
    phanloai = xeploai
    
    # Tổng hợp xếp loại học sinh và in ra màn hình
    b = zip(hocsinh, xeploai)
    b = dict(b)
    print(b)
        
    #đóng file diem_trungbinh.txt
    data.close()
    
# Khởi tạo hàm xeloai_hocsinh_thidaihoc
def xeploai_hocsinh_thidaihoc(link_dghs, phanloai, bangdiem_danhgia, hocsinh2):
    
    # Khởi tạo hàm tính tổng điểm 3 môn cho các khối A, A1, B, C
    def khoi3mon(a,b,c):
        return a + b + c 
    
    # Khởi tạo hàm tính tổng điểm 3 môn cho khối D
    def khoi3monD(a,b,c):
        return a + b + c*2
        
    # Khởi tạo list và thêm dữ liệu vào list cho các khối A, A1, B, C, D
    khoiA = []
    khoiA1 = []
    khoiB = []
    khoiC = []
    khoiD = []
    for i in bangdiem_danhgia:
        khoiA.append(i[0])
        khoiA.append(i[1])
        khoiA.append(i[2])
    
        khoiA1.append(i[0])
        khoiA1.append(i[1])
        khoiA1.append(i[5])
    
        khoiB.append(i[0])
        khoiB.append(i[2])
        khoiB.append(i[3])
    
        khoiC.append(i[4])
        khoiC.append(i[6])
        khoiC.append(i[7])
    
        khoiD.append(i[0])
        khoiD.append(i[4])
        khoiD.append(i[5])

    khoiA = [khoiA[i:i+3] for i in range(0, len(khoiA), 3)]
    khoiA1 = [khoiA1[i:i+3] for i in range(0, len(khoiA1), 3)]
    khoiB = [khoiB[i:i+3] for i in range(0, len(khoiB), 3)]
    khoiC = [khoiC[i:i+3] for i in range(0, len(khoiC), 3)]
    khoiD = [khoiD[i:i+3] for i in range(0, len(khoiD), 3)]

    loaiA = []
    for i in khoiA:
        if (khoi3mon(i[0],i[1],i[2]) >= 24):
            loaiA.append(1)
        elif (18 <= khoi3mon(i[0],i[1],i[2]) < 24):
            loaiA.append(2)
        elif (12 <= khoi3mon(i[0],i[1],i[2]) < 18):
            loaiA.append(3)
        else:
            loaiA.append(4)

    loaiA1 = []
    for i in khoiA1:
        if (khoi3mon(i[0],i[1],i[2]) >= 24):
            loaiA1.append(1)
        elif (18 <= khoi3mon(i[0],i[1],i[2]) < 24):
            loaiA1.append(2)
        elif (12 <= khoi3mon(i[0],i[1],i[2]) < 18):
            loaiA1.append(3)
        else:
            loaiA1.append(4)

    loaiB = []
    for i in khoiB:
        if (khoi3mon(i[0],i[1],i[2]) >= 24):
            loaiB.append(1)
        elif (18 <= khoi3mon(i[0],i[1],i[2]) < 24):
            loaiB.append(2)
        elif (12 <= khoi3mon(i[0],i[1],i[2]) < 18):
            loaiB.append(3)
        else:
            loaiB.append(4)

    loaiC = []
    for i in khoiC:
        if (khoi3mon(i[0],i[1],i[2]) >= 21):
            loaiC.append(1)
        elif (15 <= khoi3mon(i[0],i[1],i[2]) < 21):
            loaiC.append(2)
        elif (12 <= khoi3mon(i[0],i[1],i[2]) < 12):
            loaiC.append(3)
        else:
            loaiC.append(4)

    loaiD = []
    for i in khoiD:
        if (khoi3monD(i[0],i[1],i[2]) >= 32):
            loaiD.append(1)
        elif (24 <= khoi3monD(i[0],i[1],i[2]) < 32):
            loaiD.append(2)
        elif (20 <= khoi3monD(i[0],i[1],i[2]) < 24):
            loaiD.append(3)
        else:
            loaiD.append(4)
                
    # Tạo list xeploai_thidaihoc để gộp các list trên lại
    xeploai_thiDH = zip(loaiA, loaiA1, loaiB, loaiC, loaiD)
    xeploai_thiDH = list(xeploai_thiDH)
    # Chuyển dữ liệu tuple sang list trong danh sach con xeploai_thiDH
    xeploai_thiDH = list(map(list,xeploai_thiDH))
        
    # Tạo list xl_hs_thiDH  để gộp dữ liệu của 2 list hocsinh, xeploai_thiDH
    xl_hs_thiDH = zip(hocsinh2, xeploai_thiDH)
    xl_hs_thiDH = dict(xl_hs_thiDH)
    print(xl_hs_thiDH)
        
    # Tạo list g để gộp dữ liệu của 3 list hocsinh, phanloai, xeploai_thiDH để in ra file
    g = zip(hocsinh2, phanloai, xeploai_thiDH)
    g = list(g)
        
    # Khởi tạo 1 chuỗi tiêu đề cho dữ liệu
    tieude_file = 'Ma HS, xeploai_TB chuan, xeploai_A, xeploai_A1, xeploai_B, xeploai_C, xeploai_D'
        
    # Ghi dữ liệu ra file danhgia_hocsinh.txt
    f = open(link_dghs, "w")
    # ghi tiêu đề
    f.write(tieude_file)
    f.write("\n")
        
    # Ghi dữ liệu đọc từ list g
    for i in g:
        i = str(i)
        i = i.replace("(","")
        i = i.replace(")","")
        i = i.replace("[","")
        i = i.replace("]","")
        i = i.replace(",",";")
        i = i.replace("'","")
        f.write(i)
        f.write("\n")
     # đóng file danhgia_hocsinh.txt sau khi đã hoàn tất ghi dữ liệu
    f.close()
    
# Khởi tạo hàm Main
def Main():
    
    # gọi 2 hàm để chạy
    xeploai_hocsinh("diem_trungbinh.txt")
    xeploai_hocsinh_thidaihoc("danhgia_hocsinh.txt", phanloai, bangdiem_danhgia, hocsinh2)

# Chay ham Main
Main()
```
