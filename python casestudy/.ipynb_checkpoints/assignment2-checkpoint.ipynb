```python
while True:
    frname = input('Enter the file name to read: ')
    try:
        frhand = open(frname + '.txt', "r")
    except:
        print('File cannot be opened:', frname)
        break
    else:
        print("Successfully opened ", frname)
        print("**** ANALYZING ****")

    data = frhand.readlines()

    # Thẩm định dữ liệu
    data_idsai = []
    data_thieu = []
    data_thua = []
    data_chuan = []

    count1 = 0
    count2 = 0
    count3 = 0
    count4 = 0

    for line in data:
        line = line.rstrip()
        line = line.split(",")
        if (len(line[0]) != 9) or (line[0][1:].isnumeric() == False) or (line[0][0] != 'N'):
            data_idsai.append(line)
            count1 += 1
        else:   
            if len(line)<26:
                data_thieu.append(line)
                count2 += 1
            elif len(line)>26:
                data_thua.append(line)
                count3 += 1
            else:
                data_chuan.append(line)
                count4 += 1

    if count1 + count2 + count3 != 0:
        print("Invalid line of data: N# is invalid")
        for i in data_idsai:
            print(",".join(i))

        print("Invalid line of data: does not contain exactly 26 values (smaller 26 values)")
        for i in data_thieu:
            print(",".join(i))

        print("Invalid line of data: does not contain exactly 26 values (greater 26 values)")
        for j in data_thua:
            print(",".join(j))
    else:
        print("No errors found!")

    print("**** REPORT ****")
    print("Total valid lines of data:", count4)
    print("Total invalid lines of data:", count1 + count2 + count3)

    # Tạo list Mã học sinh và list bảng điểm 
    data_chuan_grade = []
    data_MaHS = []
    for i in data_chuan:
        data_MaHS.append(i[0])
        data_chuan_grade.append(i[1:])
    # print(data_chuan_grade)
#     print(data_MaHS)

    # bảng điểm chuẩn để so sanh đưa ra kết quả
    answer_key = ['B','A','D','D','C','B','D','A','C','C','D','B','A','B','A','C','B','D','A','C','A','A','B','D','D']            

    # Chuyển đổi điểm chữ sang điểm số thập phân
    a = []
    b = []
    for grade in data_chuan_grade:
        a.append([4 for i, j in zip(grade, answer_key) if i == j]) 
        b.append([i for i, j in zip(grade, answer_key) if i != j])         

    for i in b:
        for j in range(len(i)):
            if i[j] != "":
                i[j] = -1
            else:
                i[j] = 0

    # Ghép dữ liệu 2 list a và b thành 1 list và đưa về định dạng chuẩn d
    c = list(zip(a,b))
    d = []
    for i in c:
        for j in i:
            for k in j:
                d.append(k)

    d = [d[x:x+25] for x in range(0, len(d), 25)]
    # print(d)

    data_grade_num = []
    for i in d:
        data_grade_num.append(sum(i))

    data_grade_write = data_grade_num.copy()
#     print(data_grade_write)
    
    # 3.1 Số lượng học sinh đạt điểm cao trên 80%
    count_diemtren80 = 0
    for i in data_grade_num:
        if i > 80:
            count_diemtren80 += 1
    print("Total student of high scores:", count_diemtren80)
    # 3.2 Giá trị điểm trung bình
    print("Mean (average) score:", round(sum(data_grade_num)/len(data_grade_num), 3))
    # 3.3 Điểm cao nhất
    print("Highest score:", max(data_grade_num))
    # 3.4 Điểm thấp nhất
    print("Lowest score:", min(data_grade_num))
    # 3.5 Miền giá trị của điểm
    print("Range of scores:", max(data_grade_num) - min(data_grade_num))
    # 3.6 Giá trị trung vị
    data_grade_num.sort()
    median = None
    if (len(data_grade_num)%2 == 0):
        median = (data_grade_num[(len(data_grade_num)//2)-1] + data_grade_num[len(data_grade_num)//2])/2
    else:
        median = data_grade_num[len(data_grade_num)//2]
    print("Median score:", median)

    #  3.7 Trả về các câu hỏi bị học sinh bỏ qua nhiều nhất theo thứ tự: 
    # số thứ tự câu hỏi - số lượng học sinh bỏ qua -  tỉ lệ bị bỏ qua
    list_grade_empty = []
    for i in data_chuan_grade:
        for j in range(len(i)):
            if i[j] == "":
                list_grade_empty.append(j)
    # print(list_grade)

    dict_grade_empty = {}
    for i in list_grade_empty:
        if i not in dict_grade_empty:
            dict_grade_empty[i] = 1
        else:
            dict_grade_empty[i] += 1
    # print(dict_grade_empty)
    Max_empty = max(dict_grade_empty.values())
    list_max_empty = [k for k, v in dict_grade_empty.items() if v == Max_empty]
    list_max_empty.sort()
    print("Question that most people skip:", end = " ")
    for i in list_max_empty:
        print(str(i+1) + '-' + str(Max_empty) + '-' + str(round(Max_empty/len(data_chuan_grade), 3)), end = "; ")

    #  3.8. Trả về các câu hỏi bị học sinh sai qua nhiều nhất theo thứ tự: 
    #    số thứ tự câu hỏi - số lượng học sinh trả lời sai - tỉ lệ bị sai
    list_grade_fail = []
    for grade in data_chuan_grade:
        list_grade_fail.append([i for i, j in zip(range(len(grade)), range(len(answer_key))) if (grade[i] != answer_key[j] and grade[i]!= "")])

    list_grade_fail1 = []
    for i in list_grade_fail:
        for j in i:
            list_grade_fail1.append(j)

    dict_grade_fail = {}
    for i in list_grade_fail1:
        if i not in dict_grade_fail:
            dict_grade_fail[i] = 1
        else:
            dict_grade_fail[i] += 1

    Max_fail = max(dict_grade_fail.values())
    list_max_fail = [k for k, v in dict_grade_fail.items() if v == Max_fail]
    list_max_fail.sort()
    print("\nQuestion that most people skip:", end = " ")
    for i in list_max_fail:
        print(str(i+1) + '-' + str(Max_fail) + '-' + str(round(Max_fail/len(data_chuan_grade), 3)), end = "; ")
    
    ## Task 4: ghi điểm học sinh ra file
    data_grade_file = list(zip(data_MaHS, data_grade_write))
#     print(data_grade_file)
    fwname = input("\nEnter the file name to write:")
    fwhand = open(fwname + '.txt', "w")
    for i in data_grade_file:
        i = str(i)
        i = i.replace("(", "")
        i = i.replace(")", "")
        i = i.replace("'", "")
        fwhand.write(i+"\n")
    print("successful to write a file")
    fwhand.close()
    frhand.close()
```
