```python
## Dựa vào hàm đã viết ở bài trên, viết chương trình hỏi người dùng số lượng cần mua 3 sản phẩm sau:
## Cơm sườn - 30000
## Cơm gà - 35000
## Cơm trứng - 25000
## Sau khi người dùng đã nhập số lượng, tính và in ra tổng toàn bộ số tiền cần phải trả

def muahang(s,a):
    print('Mặt hàng',s)
    sl = int(input('Nhập vào số lượng muốn mua:'))
    return sl * a
com_ga = muahang('Cơm gà',35000)
com_suon = muahang('Cơm sườn',30000)
com_trung = muahang('Cơm trứng',25000)
print('Tổng số tiền phải trả là', com_ga + com_suon + com_trung)
```

Viết hàm kiểm tra số nguyên tố

```python
n = int(input())

def soNT(n):
    flag = True
    if(n<2):
        flag = False
    elif(n == 2):
        flag = True
    else:
        for i in range(3, (n//2) +1):
            if(n%i == 0):
                flag = False
            else: 
                flag = True
    return flag

soNT(n)
```

Viết hàm đếm số lượng từ của một chuỗi đưa vào

```python
def demtu(chuoi):
    chuoi += " "
    count = 0
    chuoi_luu = ""
    for i in chuoi:
        if i != " ":
            chuoi_luu += i
        else:
            count += 1
            chuoi_luu = ""
    return count
demtu("hoc lap trinh rat thu vi")
```

Viết hàm lấy tên đệm

Ví dụ: Nhập vào "Nguyễn Tâm Điền"

Kết quả in ra là: "Tâm"

```python
def laytendem(hoten):
    a = hoten.find(" ")
    hoten = hoten[a+1:]
    a = hoten.rfind(" ")
    hoten = hoten[:a]
    return hoten
laytendem("bui nguyen mai phuong")
```

```python
def tendem(a):
    k = 0
    p = -1
    for i in a:
        if i == " ":
            b = a[k+1:]
            break
        k += 1
    for i in b[::-1]:
        if i == " ":
            c = b[:p]
            break
        p -= 1
    return c
print(tendem("bui nguyen mai phuong"))
```

Viết một hàm vẽ hình tròn, một hàm vẽ hình vuông và một hàm vẽ hình tam giác đều (giá trị đầu vào là a)

Tất cả hàm này, bắt đầu từ 1 vị trí, sau khi vẽ ra, con trỏ sẽ tiến tới vị trí tiếp theo, cách vì đầu một khoảng a + 10 theo chiều x từ trái sang

```python
import turtle
t = turtle.Turtle()

def vehinhtron(a): # a la duong kinh
    t.penup()
    t.fd(a/2)
    t.pendown()
    t.circle(a/2)
    t.penup()
    t.fd(a/2 + 10)
    t.pendown()

def vehinhvuong(a):
    for i in range (4):
        t.fd(a)
        t.lt(90)
    t.penup()
    t.fd(a + 10)
    t.pendown()


def vehinhtamgiacdeu(a):
    for i in range (3):
        t.fd(a)
        t.lt(120)
    t.penup()
    t.fd(a + 10)
    t.pendown()
    
vehinhtron(50)  
vehinhvuong(50)
vehinhtamgiacdeu(50)
vehinhtron(50) 
vehinhvuong(50)
vehinhtamgiacdeu(50)
```
